<h1>Identyfikacja choroby Parkinsona</h1>
<h2>na podstawie mowy (Parkinson's Data Set)</h2>

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import os, sys
```

```python
data_set = pd.read_csv('parkinsons.data' , sep=",")
data_set.head()
```

<h3>Dane wejściowe </h3>

Dane wejściowe pochodzą z Parkinson’s Data Set ze strony https://archive.ics.uci.edu/ml/datasets.php

Zestaw danych został stworzony przez Maxa Little'a z Uniwersytetu Oksfordzkiego we współpracy z National Center for Voice and Speech w Denver w stanie Kolorado, który zarejestrował sygnały mowy. W oryginalnym badaniu opublikowano metody ekstrakcji cech ogólnych zaburzeń głosu.


<h4>Informacje o zbiorze danych</h4>

Zbiór składa się z pomiarów glosu 31 osób z czego 23 chorych na Parkinsona. Każda kolumna jest to konkretna miara głosy, a każdy wiersz to jedno ze 195 nagrań pobranych od badanych osób. Zbiór posiada kolumnę status, która informuje nas czy dane nagrania należą do osoby chorej (1) lub zdrowiej (0). Podczas naszej klasyfikacji będziemy właśnie chcieli przewidzieć wartość dla tej kolumny.


<h4>Kolumny </h4>
<ul>
    
<li>name - imię podmiotu badania i numer nagrania </li>
<li>MDVP:Fo(Hz) - Średnia częstotliwość podstawowa głosu</li>
<li>MDVP:Fhi(Hz) - Maksymalna częstotliwość podstawowa głosu</li>
<li>MDVP:Flo(Hz) - Minimalna częstotliwość podstawowa głosu</li>
<li>MDVP:Jitter(%),MDVP:Jitter(Abs),MDVP:RAP,MDVP:PPQ,Jitter:DDP - miary zmienności cęstotliwości podstawowej</li>
<li>MDVP:Shimmer,MDVP:Shimmer(dB),Shimmer:APQ3,Shimmer:APQ5,MDVP:APQ,Shimmer:DDA - miary zmienności amplitudy</li>
<li>NHR,HNR - Dwie miary stosunku szumu do składowych tonalnych głosu</li>
<li>status - Stan zdrowia pacjenta (1) - chory Parkinsona, (0) - zdrowy</li>
<li>RPDE,D2 - Dwie nieliniowe miary dynamicznej złożonościs</li>
<li>DFA - Wykładnik skali fraktalnej sygnału</li>
<li>spread1,spread2,PPE - Trzy nieliniowe miary podstawowej zmienności częstotliwości</li>

</ul>


<h4>Analiza danych wejściowych oraz ich wizualne przedstawienie</h4>

```python
print(data_set.shape)
```

Jak już wcześniej było wspomniane zbiór składa się z 195 wierszy, oraz z 24 wcześniej wymienionych kolumn.

```python
corr_map=data_set.corr()
sb.heatmap(corr_map)
```

Na powyższej heatmapie możemy zobaczyć korelację atrybutów naszego zbioru danych, spróbujemy uporządkować je od najmocniej skorelowanych z atrybutem status, będą to atrybuty, które najbardziej nas interesują podczas klasyfikacji.

```python
k=24

# znajdujemy kolumny które są najmocniej skorelowane z wartościami w kolumnie status
cols=corr_map.nlargest(k,'status')['status'].index

# wartości korelacji dla kolumn
coff_values=np.corrcoef(data_set[cols].values.T)

sb.heatmap(coff_values,cbar=True,annot=True,fmt='.2f',
           annot_kws={'size': 5},yticklabels=cols.values,xticklabels=cols.values)
plt.figure(figsize=(30, 30))
plt.show()
```

Widzimy już pewne zależności, jednakże aby mieć pewność wypiszemy korelację wszystkich atrybutów, w ten sposób będziemy w łatwiejszy sposób mogli wybrać te które na interesują.

```python
correlation_values=data_set.corr()['status']
correlation_values.abs().sort_values(ascending=False)
```

Mamy wartości korelacji dla każdego atrybutu, przyjmiemy więc że pominiemy dane poniżej wskaźnika korelacji 0.3, a więc porzucimy wszystkie dane poniżej atrubutu RPDE.

```python
status = data_set['status']
data = data_set[['spread1', 'PPE', 'spread2', 'MDVP:Fo(Hz)', 'MDVP:Flo(Hz)', 'MDVP:Shimmer',
                'MDVP:APQ' , 'HNR', 'Shimmer:APQ5', 'MDVP:Shimmer(dB)', 'Shimmer:APQ3', 
                'Shimmer:DDA', 'D2', 'MDVP:Jitter(Abs)', 'RPDE' ]]
data.head()
```

```python
data.describe()
```

Powyżej możemy zauważyć dosyć duży rozrzut danych, dla niektórych atrybutów wartości wyrażone są w setkach dla innych w okolicach zaledwie stutysięcznych części. Taka sytuacja skłania nas do standaryzacji danych. Jednakże dla porównania dokonamy klasyfikacji dla surowych danych jak i dla danych znormalizowanych. 

```python
y = status.values
x = data.values

scale= StandardScaler()
s_x = scale.fit_transform(x)
```

```python
train_data, test_data, \
train_target, test_target = train_test_split(x,y, test_size=0.1, random_state=10)


s_train_data, s_test_data, \
s_train_target, s_test_target = train_test_split(s_x,y, test_size=0.1, random_state=10)
```

```python
logistic_regression = LogisticRegression()
logistic_regression.fit(train_data, train_target)

s_logistic_regression = LogisticRegression()
s_logistic_regression.fit(s_train_data, s_train_target)
```

```python
id=9
print("------Dane w podstawowej formie----------")
prediction = logistic_regression.predict(test_data[id,:].reshape(1,-1))
print("Przewidziana wartość przez model dla pacjenta {0} wynosi {1}".format(id, prediction))
print("prawdziwa wartość dla tego samego pacjenta {0} to {1}".format(id, test_target[id]))


print("------Dane znormalizowane---------")
s_prediction = s_logistic_regression.predict(s_test_data[id,:].reshape(1,-1))
print("Przewidziana wartość przez model dla pacjenta {0} wynosi {1}".format(id, s_prediction))
print("prawdziwa wartość dla tego samego pacjenta {0} to {1}".format(id, s_test_target[id]))
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(test_target, logistic_regression.predict(test_data))
print("Dokładność modelu dla nieustandaryzowanych danych wynosi {0:0.2f}".format(acc))

s_acc = accuracy_score(s_test_target, s_logistic_regression.predict(s_test_data))
print("Dokładność modelu dla ustandaryzowanych danych wynosi {0:0.2f}".format(s_acc))
```

```python
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(test_target, logistic_regression.predict(test_data))
print(conf_matrix)

s_conf_matrix = confusion_matrix(s_test_target, s_logistic_regression.predict(s_test_data))
print(s_conf_matrix)
```

<h3>Analiza wyniku </h3>
Na podstawie powyższych wyników, uzyskanej dokładności wyników oraz tablicy konfuzji możemy śmiało stwiredzić, że nasz model jest dobrze wytrenowany. Dokładność modelu na poziomie 90% jest bardzo zadawalającym wynikiem. 

<h3> Porównanie klasyfikatorów dla różnych parametrów</h3>
Postaramy się znaleźć najlepszy model przy pomocy gridsearch, porównamy dzięki temu kilka algorytmów rozwiązywania regresji logistycznej a także maksymalnej ilości iteracji w celu znalezienia najlepszego połączenia dla naszego modelu.


```python
from sklearn.model_selection import GridSearchCV

parameters=[{'solver':['newton-cg', 'lbfgs', 'liblinear','sag', 'saga'], 'max_iter':[1,5,7,10,25,50,75,100]}]
model = GridSearchCV(LogisticRegression(), parameters, cv =3)
model.fit(s_x,y)
print(model.best_params_)
model.best_estimator_
```

Znaleźliśmy najlepszy model, a teraz przetestujemy go i sprawdzimy jego dokładność.
